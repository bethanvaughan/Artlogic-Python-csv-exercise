```python
import csv
import pandas as pd
import os
from csv import DictReader
```

```python
def csv_to_rows(filepath):
    
    with open(filepath, 'r') as read_obj:
        dict_reader = DictReader(read_obj)
        list_of_dict = list(dict_reader)
        return list_of_dict  
```

```python
def rows_to_csv(filepath, list_of_dictionaries, excel_file_name, output_path):
    artists = csv_to_rows(filepath)
    df = pd.DataFrame.from_dict(list_of_dictionaries)
    df = df.iloc[: , 1:]
    print (df)

    os.chdir(output_path)

    df.to_excel(excel_file_name)
```

```python
def new_stock_num_generator(filepath):
    artists = csv_to_rows(filepath)    
    artist_key = "Artist"
    values_of_Artist = [a_dict[artist_key].split(' ') for a_dict in artists]

    stock_num = []
    for i in range(0, 38):
    
        artist = values_of_Artist[i]
        
        if  artist[0]:
            first_name_letter = artist[0][0]
            second_name_letter = artist[1][0]
            zeros = '00'
            num = i
        stock_num.append(first_name_letter + second_name_letter + ' ' +  zeros + str(num))
    return(stock_num)
```
